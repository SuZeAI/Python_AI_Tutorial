## <h4>Câu 9: Phân cụm dữ liệu và loại bỏ ngoại lai </h4>

### <h4>Phân cụm của sepal and petal , lọa bỏ ngoại lai</h4>

```python
# thêm thư viện matplotlib
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
df8 = df7.copy()
```

```python
# y = np.concatenate((df8['sepal.length'],df8['petal.length']), axis=0)
# x = np.concatenate((df8['sepal.width'],df8['petal.width']), axis=0)
X = np.concatenate((df8[df8.columns[0: 2]], df8[df8.columns[2 : 4]]), axis=0)
n_cluster = 3
```

```python
def kmeans_init_centers(X, n_cluster):
  # random k index beetween 0 and shape(X) without duplicate index.
  # Then return X[index] as cluster
  return X[np.random.choice(X.shape[0], n_cluster, replace=False)]
```

```python
def kmeans_predict_labels(X, centers):
  D = cdist(X, centers)
  # return index of the closest center
  return np.argmin(D, axis = 1)
```

```python
def kmeans_update_centers(X, labels, n_cluster):
  centers = np.zeros((n_cluster, X.shape[1]))
  for k in range(n_cluster):
    # collect all points assigned to the k-th cluster 
    Xk = X[labels == k, :]
    # take average
    centers[k,:] = np.mean(Xk, axis = 0)
  return centers
```

```python
def kmeans_has_converged(centers, new_centers):
  # return True if two sets of centers are the same
  return (set([tuple(a) for a in centers]) == 
      set([tuple(a) for a in new_centers]))
```

### <h4>Phân cụm</h4>

```python
# Phân cụm
def kmeans_visualize(X, centers, labels, n_cluster, title):
  plt.xlabel('weight') # label trục x
  plt.ylabel('') # label trục y
  plt.title(title) # title của đồ thị
  plt_colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'] # danh sách các màu hỗ trợ
 
  for i in range(n_cluster):
    data = X[labels == i] # lấy dữ liệu của cụm i
    plt.plot(data[:, 0], data[:, 1], plt_colors[i] + '^', markersize = 6, label = 'cluster_' + str(i)) # Vẽ cụm i lên đồ thị
    plt.plot(centers[i][0], centers[i][1],  plt_colors[i+4] + 'o', markersize = 10, label = 'center_' + str(i)) # Vẽ tâm cụm i lên đồ thị
  plt.legend() # Hiện bảng chú thích
  plt.show()
```

```python
def kmeans(init_centes, init_labels, X, n_cluster):
# thuật toán K-means
  centers = init_centes
  labels = init_labels
  times = 0
  while True:
    labels = kmeans_predict_labels(X, centers)
    # kmeans_visualize(X, centers, labels, n_cluster, 'Assigned label for data at time = ' + str(times + 1))
    new_centers = kmeans_update_centers(X, labels, n_cluster)
    if kmeans_has_converged(centers, new_centers):
      break
    centers = new_centers
    # kmeans_visualize(X, centers, labels, n_cluster, 'Update center possition at time = ' + str(times + 1))
    times += 1
  kmeans_visualize(X, centers, labels, n_cluster, 'Update center possition at time = ' + str(times + 1))
  
  return (centers, labels, times)
```

```python
def main():
    init_centers = kmeans_init_centers(X, n_cluster)
    # print(init_centers) # In ra tọa độ khởi tạo ban đầu của các tâm cụm
    init_labels = np.zeros(X.shape[0])
    # kmeans_visualize(X, init_centers, init_labels, n_cluster, 'Init centers in the first run. Assigned all data as cluster 0')
    centers, labels, times = kmeans(init_centers, init_labels, X, n_cluster)
main()
```

### <h4>Loại bỏ ngoại lai</h4>

```python
# Phân hoạch

def minus_axis(a, b):
     return a - b
def kmeans_visualize1(X, centers, labels, n_cluster, title):
  plt.xlabel('weight') # label trục x
  plt.ylabel('') # label trục y
  plt.title(title) # title của đồ thị
  plt_colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'] # danh sách các màu hỗ trợ
  for i in range(n_cluster):
    check = True
    data = X[labels == i] 
    for j in range(data.shape[0]):
          if (pow(minus_axis(data[j][0], centers[i][0]), 2) + pow(minus_axis(data[j][1], centers[i][1]), 2)) < 2 * pow(0.1, 2) and check:
              plt.plot(data[j][0], data[j][1], plt_colors[i] + '^', markersize = 6, label = 'cluster_' + str(i)) # Vẽ cụm i lên đồ thị
              plt.plot(centers[i][0], centers[i][1],  plt_colors[i+4] + 'o', markersize = 10, label = 'center_' + str(i)) # Vẽ tâm cụm i lên đồ thị
              check = False
          elif (pow(minus_axis(data[j][0], centers[i][0]), 2) + pow(minus_axis(data[j][1], centers[i][1]), 2)) < 2 * pow(0.1, 2):
              plt.plot(data[j][0], data[j][1], plt_colors[i] + '^', markersize = 6) # Vẽ cụm i lên đồ thị
              plt.plot(centers[i][0], centers[i][1],  plt_colors[i+4] + 'o', markersize = 10) # Vẽ tâm cụm i lên đồ thịị
  plt.legend() 
  plt.show()
```

```python
def kmeans1(init_centes, init_labels, X, n_cluster):
# thuật toán K-means
  centers = init_centes
  labels = init_labels
  times = 0
  while True:
    labels = kmeans_predict_labels(X, centers)
    # kmeans_visualize(X, centers, labels, n_cluster, 'Assigned label for data at time = ' + str(times + 1))
    new_centers = kmeans_update_centers(X, labels, n_cluster)
    if kmeans_has_converged(centers, new_centers):
      break
    centers = new_centers
    # kmeans_visualize(X, centers, labels, n_cluster, 'Update center possition at time = ' + str(times + 1))
    times += 1
  kmeans_visualize1(X, centers, labels, n_cluster, 'Update center possition at time = ' + str(times + 1))
  
  return (centers, labels, times)
```
