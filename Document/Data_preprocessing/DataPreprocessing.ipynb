## <h4> Câu 4: Chuẩn hóa dữ liệu theo min max</h4>

```python
df3 = df1.copy()
df3
```

```python
age_min = df3.Age.min()
age_max  = df3.Age.max()
income_min = df3.Income.min()
income_max = df3.Income.max()
```

```python
df3.Age = (df3.Age - age_min) / (age_max - age_min)
df3.Income = (df3.Income - income_min) / (income_max - income_min)
```

```python
df3
```

## <h4>Câu 6: Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính Online Shopper dựa trên các thuộc tính khác</h4>

```python
df5 = df4.copy()
df5['Online Shopper'].replace({"No": 0, "Yes": 1}, inplace=True)
df5
```

```python
x = df5.drop("Age", axis=1)
y = df5["Online Shopper"]
thread_point = int(0.7 * len(df5))
x
```

```python
# tạo tập test và train
thread_point
```

```python
x_train, x_test = x[:thread_point], x[thread_point:]
y_train, y_test = y[:thread_point], y[thread_point:]
```

```python
from sklearn.linear_model import LogisticRegression
lireg = LogisticRegression()
lireg.fit(x_train, y_train)
```

```python
print("Predicted label: {}, True label: {}".format(lireg.predict(x_test), y_test.values))
```
